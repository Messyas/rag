Recuperaçao de informaçao:

Finalidade: Achar informacao util para um contexto real e especifico de cada usuario, essa informacao em geral esta contida nos documentos, pode ser algo sobre uma doenca, um jogo ou ate uma receita de lasanha.

Pontos interessantes:
1 - Nao necessaraimente existe uma resposta certa, pois depende da necessidade do usuario naquele contexto especifico.
2 - Lida principalmente com busca de documentos ou informacoes em base de dados massivas.
3 - Surgiu para criar catalogos (sistemas de indices) computadorizados.

A "Necessidade de informacao" do usuario é representada em formas de consultas, em geral feitas em linguagem natural.
Essa escrita padrao, que muitas vezes pode tem sentido semantico ambiguo (Semântica é sobre o significado) acaba dificultando a busca ou a tornando imprecisa.

Essencialmente o retorno de um sistema de recuperacao é a MAIOR QUANTIDADE de Documentos relevantes e a MENOR QUANTIDADE de documentos irrelevantes.

O que é relevante depende do objetivo, pode ser relevancia afetiva, situacional, cognitiva, tópica e algoritmica. Alem disso, se achar um em especifico, talvez todo o resto deixe de ser relevante, entao o fator temporal tambem é relevante.

A principio a recuperacao pode ser tao intuitiva quanto comparar de "string usuario" == "string documento" (o que é chamado de full-text retrieval)
Existem varias tecnicas tradicionais nessa area, por exemplo: Classificacao do documento dentro de um conjunto de tópicos pre determinados, classificacao facetada, uso de thesaurus (organizacao por proximidade) ou tecnicas de consulta, uso de informacao da rede, como o algoritmo de PageRank, tecnicas de sistemas de recomendacao, etc.

- Tarefas de recuperacao da informacao:

1. Busca Ad-Hoc: Buscar documentos em uma colecao fixa a partir de uma consutla gerada pelo usuario.
2. Filtragem: verificar se documentos que chegam em uma coleção atendem a uma consulta previamente cadastrada por um usuário
3. Roteamento, fazer a tarefa de filtragem e ordenando os documentos por relevância
4. Browsing ou navegação, navegação entre documentos como feito na Web.

- Relevancia:
É a medida ou Grau de correspondencia ou utilidade existente entre um texto ou documento e uma consulta ou requisito de informacao para uma determinada pessoa.

Para ser melhor, o sistema de RI pode ordenar (rank) os documentos  de acordo com o maior grau de semelhanca com o que se pede pelo usuario.

É isso que dita o que vai calcular a classificacao (ordenacao) de documentos em um conjunto de respostas a uma consutla do usuario.

Os passos gerais de um sistema de recuperacao sao

1. a consulta: vao ser definidos conjuntos de termos, associados ou nao a operadores Booleanoos, que representa a necessidade de informacao do usuario.
2. Indexacao: Estruturas de dados para lgiar documentos a colecoes. Como por exemplo listas invertidas de termos ou documentos.
3. Pesquisa e Ordenacao: envolve o processo de recuperação de documentos de acordo com a consulta do usuário e sua ordenação através de um grau de similaridade entre o documento e a consulta.

Existem pelo menos 3 modelos classificos para RI:

1. Booleano
2. Vetorial
3. Probabilistico

Conceitos estruturais aparecem em praticamente qualquer sistema de busca: indexação, consulta e classificação (ranking).

Eles correspondem às três fases principais do funcionamento de um motor de busca.

Fluxo geral de um sistema de RI: documentos → indexação → índice → consulta do usuário → recuperação → classificação → resultados ordenados


1. Indexação

Indexação é o processo de organizar os documentos de forma que possam ser buscados rapidamente.

Sem indexação, um sistema teria que ler todos os documentos toda vez que uma consulta fosse feita.

Exemplo de Corpus (Corpus é uma coleção estruturada de textos utilizada para análise linguística, treinamento de modelos ou recuperação de informação):

- D1: this movie is amazing
- D2: this horror movie is terrible
- D3: amazing acting but boring story

Após pré-processamento simples:

- D1 → movie amazing
- D2 → horror movie terrible
- D3 → amazing acting boring story

O sistema cria uma estrutura chamada índice invertido.

Índice invertido:

- movie → D1, D2
- amazing → D1, D3
- horror → D2
- terrible → D2
- acting → D3
- boring → D3
- story → D3

O nome “invertido” vem do fato de inverter a relação:

- documento → palavras
- vira
- palavra → documentos

Assim, quando alguém busca por “movie”, o sistema acessa diretamente:

- movie → D1, D2

Isso torna a busca extremamente rápida.

Em sistemas maiores, o índice também guarda:

1. frequência da palavra no documento (tf)
2. posição da palavra
3. peso do termo

Consulta (Query Processing)

A consulta é o texto digitado pelo usuário.

Exemplo:

amazing movie

Antes de buscar no índice, a consulta normalmente passa pelo mesmo processamento aplicado aos documentos:

1. normalização (lowercase)
2. remoção de pontuação
3. tokenização

Consulta processada:

["amazing", "movie"]

Agora o sistema procura no índice.

1. movie → D1, D2
2. amazing → D1, D3

Candidatos encontrados:

- D1, D2, D3

Esses documentos contêm pelo menos um termo da consulta.

Mas ainda não sabemos qual é o mais relevante.

É aqui que entra a classificação.

Classificação (Ranking)

Classificação é o processo de ordenar os documentos recuperados por relevância.

Sem ranking, os resultados seriam apenas uma lista não ordenada.

Exemplo:

- consulta: amazing movie

Documentos candidatos:

- D1, D2, D3

Agora calculamos um score de relevância.

Usando um modelo de similaridade:

score(query, documento)

Exemplo de resultado:

- D1 → 0.82
- D3 → 0.45
- D2 → 0.20

Ranking final:

1. D1
2. D3
3. D2

Esse score pode ser calculado usando diferentes modelos:

1. TF-IDF + similaridade do cosseno
2. BM25
3. modelos neurais (embeddings)

Exemplo simulado dos 3 ultimos conceitos juntos:

Corpus:

D1: this movie is amazing
D2: this horror movie is terrible
D3: amazing acting but boring story

Consulta:

amazing movie

Indexação encontra:

movie → D1, D2
amazing → D1, D3

Documentos candidatos:

D1, D2, D3

Classificação calcula scores:

D1 → 0.82
D3 → 0.45
D2 → 0.20

Resultados apresentados ao usuário:

this movie is amazing

amazing acting but boring story

this horror movie is terrible
